```python
import pandas as pd
import numpy as np
import seaborn as sbn
```

```python
sam = pd.read_csv('Movie_Data.csv')
sam.head()
```

```python
manh = pd.read_csv('movies.csv')
for i in ['Genre', 'Country', 'Director', 'Writer', 'Stars']:
    manh[i] = manh[i].str[1:-1].str.replace("'", "")
```

```python
phu = pd.read_csv('Movie_data_phu.csv')
for i in ['Genre', 'Country', 'Director', 'Writer', 'Stars']:
    phu[i] = phu[i].str[1:-1].str.replace("'", "")
phu.head()
```

```python
vanh = pd.read_csv('Movie_Data1.csv', names=sam.columns)
vanh.head()
```

```python
sam.columns[~sam.columns.isin(manh.columns)]
```

```python
manh.columns.isin(sam.columns)
```

```python
sam_vanh = pd.concat([sam, vanh])
sam_vanh.head()
```

```python
sam.columns
```

```python
manh.columns
```

```python
manh['Cost'] = manh['Budget']
manh['Storyline'] = manh['Story']
manh['Income'] = manh['Gross worldwide']
```

```python
manh = manh[manh.columns[manh.columns.isin(sam.columns)]]
```

```python
sam_vanh_manh = pd.concat([sam_vanh, manh])
sam_vanh_manh.head()
```

```python
sam.columns[~sam.columns.isin(phu.columns)]
```

```python
phu.columns
```

```python
phu['Storyline'] = phu['Story']
```

```python
sam.columns[~sam.columns.isin(phu.columns)]
```

```python
phu = phu[phu.columns[phu.columns.isin(sam.columns)]]
```

```python
data = pd.concat([sam_vanh_manh, phu])
data
```

```python
data.info()
```

```python
data[(data['Name'].isnull()) | (data['Poster'].isnull()) | (data['Storyline'].isnull())]
```

```python
data = data.reset_index()
```

```python
data = data[data.columns[~data.columns.isin(['index'])]]
```

```python
data['Certificate'][data['Certificate'].isnull()] = '1'
data['Certificate'][data['Certificate'].str.contains('h')] = np.nan
data['Certificate'][data['Certificate'] == '1'] = np.nan
data['Certificate'][data['Certificate'].isnull()] = 'Unrated'
```

```python
data
```

```python
data2 = data[['Name', 'Certificate', 'Poster', 'Genre', 'Director', 'Writer', 'Stars', 'Storyline', 'Rating', 'Rating Count']]
```

```python
data2
```

```python
mask = ~data2['Name'].isnull()
for i in data2.columns:
    mask = (mask & ~data2[i].isnull())
data2 = data2[mask]
```

```python
data2.info()
```

```python
data.to_csv('Full_Raw_Data.csv')
data2.to_csv('Cleaned_Data.csv')
```
